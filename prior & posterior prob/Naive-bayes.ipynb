```python
import pandas as pd
from collections import defaultdict
```

```python
# Convert the list of lists to a DataFrame
data = [
    ['Day', 'Outlook', 'Temp.', 'Humidity', 'Wind', 'Decision'],
    ['D1', 'Sunny', 'Hot', 'High', 'Weak', 'No'],
    ['D2', 'Sunny', 'Hot', 'High', 'Strong', 'No'],
    ['D3', 'Overcast', 'Hot', 'High', 'Weak', 'Yes'],
    ['D4', 'Rain', 'Mild', 'High', 'Weak', 'Yes'],
    ['D5', 'Rain', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['D6', 'Rain', 'Cool', 'Normal', 'Strong', 'No'],
    ['D7', 'Overcast', 'Cool', 'Normal', 'Strong', 'Yes'],
    ['D8', 'Sunny', 'Mild', 'High', 'Weak', 'No'],
    ['D9', 'Sunny', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['D10', 'Rain', 'Mild', 'Normal', 'Weak', 'Yes'],
    ['D11', 'Sunny', 'Mild', 'Normal', 'Strong', 'Yes'],
    ['D12', 'Overcast', 'Mild', 'High', 'Strong', 'Yes'],
    ['D13', 'Overcast', 'Hot', 'Normal', 'Weak', 'Yes'],
    ['D14', 'Rain', 'Mild', 'High', 'Strong', 'No']
]

columns = data[0]
rows = data[1:]

df = pd.DataFrame(rows, columns=columns)

def calculate_prior_probabilities(df):
    priors = df['Decision'].value_counts(normalize=True)
    return priors.to_dict()

def calculate_likelihoods(df, feature, value, decision):
    subset = df[df['Decision'] == decision]
    likelihood = subset[feature].value_counts(normalize=True).get(value, 0)
    return likelihood

def calculate_posterior(df, conditions):
    priors = calculate_prior_probabilities(df)
    posteriors = defaultdict(float)

    for decision in priors:
        posterior = priors[decision]
        for feature, value in conditions.items():
            likelihood = calculate_likelihoods(df, feature, value, decision)
            posterior *= likelihood
        posteriors[decision] = posterior

    # Normalize to sum to 1
    total = sum(posteriors.values())
    for decision in posteriors:
        posteriors[decision] /= total

    return posteriors


def predict_decision(df, conditions):
    posteriors = calculate_posterior(df, conditions)
    # Predict the decision with the highest posterior probability
    prediction = max(posteriors, key=posteriors.get)
    return prediction

# Example usage:
new_instance = {'Outlook': 'Sunny', 'Temp.': 'Hot', 'Humidity': 'High', 'Wind': 'Weak'}
prediction = predict_decision(df, new_instance)
print("Predicted Decision:", prediction)

```
